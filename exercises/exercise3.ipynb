```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.divide(np.pi, 6)
b = np.divide(np.pi, 3)
f = lambda x: np.sin(x) * np.cos(2 * x)
f_double_derivative = lambda x: 5 * np.sin(x) - 18 * np.power(np.cos(x), 3) * np.sin(x)
interval = np.linspace(a, b, 1000)

plt.plot(interval, np.abs(f(interval)))
plt.show()

max_val = np.max(np.abs(f_double_derivative(interval)))
error_bound = np.divide(np.power(b - a, 3), 12) * max_val
print(max_val)
print(error_bound)
```

## Problem 2
### a)

```python
p1 = np.divide(
    np.sqrt(525 + np.multiply(70, np.sqrt(30))),
    35
)

p2 = np.divide(
    np.sqrt(525 - np.multiply(70, np.sqrt(30))),
    35
)

w1 = np.divide(18 - np.sqrt(30), 36)
w2 = np.divide(18 + np.sqrt(30), 36)
```

```python
points = np.array([-p1, -p2, p2, p1])
weights = np.array([w1, w2, w2, w1])
```

```python
f = lambda x: np.exp(x)
translate = lambda x, a, b: np.divide(b - a, 2) * x + np.divide(b + a, 2)
integral = lambda x, weights, a, b: np.divide(b - a, 2) * np.sum(np.multiply(weights, f(translate(x, a, b))))
```

```python
a, b = -3, 3
integral_value = integral(x=points,weights=weights, a=a, b=b)
print(round(integral_value, 5))
```

## Problem 4
### a)

```python
g = lambda x: np.power(x, 2)
m = 10
```

```python
def method(f, a, b, m):
    xs = np.linspace(a, b, m+1)
    ys = [f(x) for x in xs]
    s = (1/2) * (ys[0] + ys[-1]) + sum(ys[1:-1])
    return s
```

```python
print(method(g, 0, 1, m))
print(method(g, 0, 1, 30))
```
